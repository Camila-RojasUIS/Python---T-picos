Este valor es relativamente bajo, lo que sugiere que, en promedio, los cuadrados de las diferencias entre los valores reales y los predichos son pequeños. Sin embargo, debido a la penalización cuadrática, este valor puede estar influenciado por algunos errores grandes.

Este valor también es relativamente bajo, lo que indica que, en promedio, la diferencia absoluta entre los valores reales y los predichos es pequeña. El MAE proporciona una visión más intuitiva de la magnitud del error promedio.

**Comparación de MSE y MAE**
- MSE es ligeramente más bajo que el MAE, lo que puede sugerir que no hay errores extremadamente grandes que estén influyendo desproporcionadamente en la medida del error.
- Ambos valores bajos indican que el modelo tiene un buen rendimiento en términos de precisión de predicción.
